```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("honeyproduction.csv")
print(df.head())
```

```python
prod_per_year = df.groupby("year").totalprod.mean().reset_index()
X = prod_per_year["year"]
X = X.values.reshape(-1, 1)
print(X)
```

```python
y = prod_per_year['totalprod']
```

```python
plt.scatter(X, y)
plt.xlabel('Year')
plt.ylabel('Total production')
plt.title('Total honey production over the years')
plt.show()
```

**Create and Fit Linear Regression Model**

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
# print out slope (coefficients) and intercept of the line
print("Slope (coefficients):", regr.coef_)
print("Intercept:",regr.intercept_)
```

```python
# Making prediction using the fitting model
y_predict = regr.predict(X)
print(y_predict)
```

```python
plt.scatter(X, y, label='Actual data')
plt.plot(X, y_predict, color='red', label='Predicted line')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.title('Actual vs Predicted Total Honey Production Over the Years')
plt.legend()
plt.show()
```
